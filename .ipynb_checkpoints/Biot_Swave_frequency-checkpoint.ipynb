```python
import numpy as np
Bg=50e9
Bsk = 1e5
Bf = 2.18e9
Gsk = 0.1e9
n = 0.4
D = Bg*(1+n*(Bg/Bf-1))
C = Bg*(Bg -Bsk)/(D-Bsk)
H = Bsk +4*Gsk/3 + (Bg-Bsk)*(Bg-Bsk)/(D-Bsk)
M = Bg*Bg/(D-Bsk)

f = np.linspace(0,30000, num=30000) #frequency Hz
omega = 2*np.pi*f # angular frequncy
a = 0.12e-4# pore size parameter m
a2 = a*a
nu = 0.001 # fluid viscosity Pa.s
alpha = 1.67# tortuosity factor
K = 10e-11 #hydraulic conductivity m2  (1darcy=10e-12)


rho_f = 1000.# fluid density kg/m3
rho_g = 2644 # grain density kg/m3
n = 0.4  # porosity
rho_mix= (1-n)*rho_g + n*rho_f
```

```python
import scipy.special as sp
dseta = np.sqrt(omega*a2*rho_f/nu)

Bes_arg = dseta*np.exp(-1j*np.pi/4) #argument der Besselfunktion in T
J1=sp.jv(1,Bes_arg) # Besselfunktion 1. Ordnung
J0=sp.jv(0,Bes_arg) # Besselfunktion 0. Ordnung
Jdiv=J1/J0
T = np.exp(1j*3*np.pi/4)*Jdiv
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
fig = plt.figure()
#ax1 = plt.subplot(311)
#ax1.plot(dseta)
ax2 = plt.subplot(211)
ax2.plot(T.real)
ax3 = plt.subplot(212)
ax3.plot(T.imag)
```

```python
F1 = dseta*T/4
F2 = dseta*T/4/(1+2j*T/dseta)
denom = (1+2j*T/dseta)
import matplotlib.pyplot as plt
%matplotlib notebook
fig = plt.figure()
#ax1 = plt.subplot(311)
#ax1.plot(dseta)
ax2 = plt.subplot(111)
ax2.plot(F1.real)
ax2.plot(F2.real, 'r')
#ax2.plot(denom, 'g')
#ax3 = plt.subplot(212)
#ax3.plot(F1.imag)
```

```python
q1 = alpha*rho_f/n
q2 = 1j*nu*F2/omega/K
q = q1 - q2

%matplotlib notebook
fig3 = plt.figure()
ax = plt.subplot(211)
ax.plot(q.real)
ax = plt.subplot(212)
ax.plot(q.imag)
```

```python
Vs2 = Gsk*q/(rho_mix*q - rho_f*rho_f)

Vs = np.sqrt(Vs2)
%matplotlib notebook
fig = plt.figure()
ax = plt.subplot(211)
ax.plot(Vs.imag)
ax2= plt.subplot(212)
ax2.plot(Vs.real)
```
